```python
import pandas as pd
import numpy as np

# Gerar dados sintéticos para incidentes de cibersegurança em Angola
data = {
    'Ano': np.random.choice(range(2021, 2026), 100),
    'Setor': np.random.choice(['Bancos', 'Telecomunicações', 'Governo', 'Educação', 'Saúde', 'Energia'], 100),
    'Tipo_Ataque': np.random.choice(['Phishing', 'Malware', 'Ransomware', 'Engenharia Social', 'DDoS', 'Zero-Day'], 100),
    'Impacto': np.random.choice(['Roubo de Dados', 'Fraudes', 'Interrupção', 'Perda Financeira', 'Reputação', 'Sabotagem'], 100),
    'Vetor': np.random.choice(['Web', 'Email', 'Rede Interna', 'IoT', 'Cloud'], 100),
    'Numero_Ataques': np.random.randint(100, 5000, 100),
    'Percentagem_Impacto': np.random.randint(5, 95, 100)
}

df_ciberseguranca_angola = pd.DataFrame(data)

print("Template DataFrame para Incidentes de Cibersegurança em Angola criado:")
display(df_ciberseguranca_angola.head())
```

```python
df_ataques_por_tipo = df_ciberseguranca_angola.groupby('Tipo_Ataque')['Numero_Ataques'].mean().reset_index()
print("Média de ataques por Tipo_Ataque:")
display(df_ataques_por_tipo)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Tipo_Ataque', y='Numero_Ataques', data=df_ataques_por_tipo, palette='viridis', hue='Tipo_Ataque', legend=False)
plt.title('Média de Ataques por Tipo de Ataque')
plt.xlabel('Tipo de Ataque')
plt.ylabel('Média de Número de Ataques')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df_ataques_por_setor = df_ciberseguranca_angola.groupby('Setor')['Numero_Ataques'].sum().reset_index()
print("Número total de ataques por Setor:")
display(df_ataques_por_setor)
```
